```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
my_url = 'https://www.thehindu.com/sci-tech/technology/'
page_soup = soup(uReq(my_url).read(),"html.parser")
headings=page_soup.findAll("h3") 
l1=[]
l2=[]
for i in range (0,len(page_soup.findAll("h3"))):
    print(headings[i].text)
    try:
        print(headings[i].a['title'])
        l1.append(headings[i].text)
        l2.append(headings[i].a['title'])
    except:
        print('date not specified')
    print('___________________________________________________')
print(len(page_soup.findAll("h3")))
s1=json.dumps(l1,indent=4)
with open('l1.json','w') as f1:
    json.dump(s1,f1)
print(s1)
s2=json.dumps(l2,indent=4)
with open('l2.json','w') as f2:
    json.dump(s2,f2)
print(s2)
```

```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
my_url = 'https://www.thehindu.com/sci-tech/technology/'
l3=[]
page_soup = soup(uReq(my_url).read(),"html.parser")
headings=page_soup.findAll("h3") 
for i in range (0,len(page_soup.findAll("h3"))):
    try:
        print(headings[i].a['href'])
        l3.append(headings[i].a['href'])
    except:
        print('error 404')
    print('_______________________________________________________________________________________________________________')  
print(l3)
s3=json.dumps(l3,indent=4)
with open('l3.json','w') as f3:
    json.dump(s3,f3)
print(s3)
```

```python
from bs4 import BeautifulSoup
import requests
import json
l4=[]
for j in range(len(l3)):
    my_url = requests.get(l3[j]).text
    soup = BeautifulSoup(my_url,"lxml")
    s=soup.find_all('p')
    for i in range(len(s)):
        l4.append(s[i].text)
#Only print in json format
s4=json.dumps(l4,indent=4)
with open('l4.json','w') as f4:
    json.dump(s4,f4)
print(s4)
```

```python
from bs4 import BeautifulSoup
import requests
import json
l5=[]
my_url = requests.get('https://www.thehindu.com/sci-tech/technology/internet/google-rolls-out-updated-privacy-tools-for-youtube-maps/article29575336.ece').text
soup = BeautifulSoup(my_url,"lxml")
link = soup.find_all('img')
for i in range(len(link)):
    print(link[i]['src'])
    l5.append(link[i]['src'])
s5=json.dumps(l5,indent=4)
with open('l5.json','w') as f5:
    json.dump(s5,f5)
print(s5)
```
